```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Ans:Multiple inheritance occurs when a class inherits attributes and methods from more than one parent class. It can be a powerful tool, but it can also lead to complexities, especially when there are method name conflicts.

Python Code Example:

Python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass

class Flyable:
    def fly(self):
        print("I can fly!")

class Bird(Animal, Flyable):
    def __init__(self, name):
        super().__init__(name)

    def speak(self):
        print(f"{self.name} says tweet!")

# Create a bird object
bird = Bird("Sparrow")
bird.speak()  # Output: Sparrow says tweet!
bird.fly()    # Output: I can fly!
```

```python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

Ans:  Getters and setters are methods used in object-oriented programming to control access to an object's attributes. While Python doesn't strictly enforce private attributes like some other languages, they are often used for data encapsulation and to add logic around getting and setting values.

A Python Example
Python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    # Getter for name
    @property
    def name(self):
        return self._name

    # Setter for name
    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise TypeError("Name must be a string")
        self._name = value

    # Getter for age
    @property
    def age(self):
        return self._age

    # Setter for age
    @age.setter
    def age(self, value):
        if not isinstance(value, int):
            raise TypeError("Age must be an integer")
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Ans: Method overriding occurs when a subclass provides a specific implementation of a method that is already defined in its parent class. The method in the subclass must have the same name and parameters as the method in the parent class.   

Example:

Python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")   


# Create objects
animal = Animal()
dog = Dog()
cat = Cat()

# Call the speak method for each object
animal.speak()  # Output: Animal speaks
dog.speak()     # Output: Dog barks
cat.speak()     # Output: Cat meows
```
